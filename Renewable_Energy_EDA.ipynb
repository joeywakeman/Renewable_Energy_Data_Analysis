# Exploratory Data Analysis Part 1 - Renewable Energy Around the Globe in 2019

### Global analysis of the most recent, full year of renewable energy data

```python
df_current_stats = df3[df3["Year"] == 2019]
df_current_stats.head()
```

## Visualizing access to electricity and clean energy usage for every country in 2019

```python
stripplot = sns.stripplot(data = df_current_stats[["Access_to_Electricity(%_Pop)",
                                                   "Percent_Clean_Energy"]],
                                                   palette = "YlGnBu", jitter=.35, 
                                                   zorder=1)

labels = ("Percent Population with \nAccess to Electricity",
          "Clean Energy Share\n of All Energy Consumed")
stripplot.xaxis.set_ticks(range(len(labels)))
stripplot.set_xticklabels(labels)

plt.title("Countries of the World: 2019 Energy Data")
plt.rcParams["figure.figsize"] = (5,5)

plt.show()
```

Each dot on the chart above represents data from one individual country in the year 2019. Most countries have widespread access to electricity, but electricity is sourced from clean and unclean sources.

## Importing and merging a dataset that lists countries with their corresponding continents

```python
df_continents = pd.read_csv(r"/Users/josephwakeman/Downloads/continents3.csv")
df_continents.head()
```

```python
df_continents1 = df_continents.rename(columns = {"country":"Country", 
                     "continent":"Continent",
                     "sub_region":"Sub_Region"})
df_continents1.head()
```

## Charting the top 50 countries with the highest clean energy share in the world

```python
df_topfifty = df_current_stats_w_region[["Country","Percent_Clean_Energy",
                                         "Continent"]].sort_values(by = "Percent_Clean_Energy", 
                                                                   ascending = False).head(50)
df_topfifty.head()           
```

Comments: Africa clearly leads the world in clean energy share dominance, and the nordic countries are the only European countries in the top 50.

## Charting the 50 countries with the lowest clean energy share in the world

```python
df_lowfifty = df_current_stats_w_region[["Country","Percent_Clean_Energy",
                                         "Continent"]].sort_values(by = "Percent_Clean_Energy", 
                                                                   ascending = True).head(50)
df_lowfifty.head()           
```

```python
hue_order = ["Africa","Asia","Europe","Americas","Oceania"]
bar1 = sns.barplot(data=df_lowfifty, x="Country", y="Percent_Clean_Energy", 
                   hue="Continent", palette="tab10", dodge=False, hue_order=hue_order)
plt.rcParams["figure.figsize"] = (15,5)
plt.xticks(rotation=90)
bar1.set_ylabel("Percent Clean Energy of Energy Consumed")
plt.title("The 50 Countries with the Lowest Clean Energy Share of Energy Consumed - 2019")
plt.show()
```

## Exploring correlations in 2019

```python
df_current_correlations = df_current_stats_w_region[["Access_to_Electricity(%_Pop)",
                                                     "Renewable_Energy_Generating_Capacity/Capita",
                                                     "Percent_Clean_Energy",
                                                     "Energy_Consumption/Capita(kWh)",
                                                     "GDP_Per_Capita",
                                                     "Continent"]]
df_current_correlations.head()
```

```python
sns.heatmap(df_current_correlations.select_dtypes("float").corr(), annot = True, cmap = "vlag", 
            xticklabels = ["Access to Electricity", 
                           "Renewable Energy Generating Capacity", 
                           "Percent Clean Energy", 
                           "Energy Consumption per Capita", 
                           "GDP per Capita"],
            yticklabels = ["Access to Electricity", 
                           "Renewable Energy Generating Capacity", 
                           "Percent Clean Energy", 
                           "Energy Consumption per Capita", 
                           "GDP per Capita"])

plt.title("Renewable Energy Correlations in 2019")
plt.rcParams["figure.figsize"] = (7,5)
plt.show()
```

Comments: There are a variety of strong and weak correlations in our dataset. The strongest positive correlations are between energy consumption per capita, GDP per capita, and access to electricity. The strongest negative correlations are between access to electricity and percent clean energy, and access to electricity and energy consumption per capita.

Wealthy nations seemingly have widespread access to electricity, use a high amount of energy, generate a high amount of energy, and rely mostly on unclean energy. Less wealthy nations have less access to electricity, use less energy, generate less energy, and rely more on renewable energy.

Interestingly, renewable energy generating capacity is uncorrelated with clean energy share of energy consumed.

## Plotting GDP per Capita vs Energy Consumption

```python
df_gdp_and_consumption = df_current_correlations[["Continent",
                                                  "Energy_Consumption/Capita(kWh)",
                                                  "GDP_Per_Capita"]]
df_gdp_and_consumption.head()
```

```python
hue_order = ["Africa","Asia","Europe","Americas","Oceania"]

jointplot = sns.jointplot(data=df_gdp_and_consumption, x="GDP_Per_Capita", 
                          y="Energy_Consumption/Capita(kWh)", 
                          hue="Continent", palette="tab10", hue_order=hue_order)
regplot = sns.regplot(data=df_gdp_and_consumption, x="GDP_Per_Capita", 
                      y="Energy_Consumption/Capita(kWh)", 
                      scatter=False, ax=jointplot.ax_joint)

plt.ylim(0,220000)
plt.xlim(0,120000)

regplot.set_xlabel("GDP per Capita (Thousands of Dollars)")
regplot.set_ylabel("Energy Consumption per Capita (Thousands of kWh)")

regplot.ticklabel_format(axis="y",scilimits=(3,3))
regplot.ticklabel_format(axis="x",scilimits=(3,3))

plt.show()
```

Comments: Most countries, especially in Africa, have relatively low GDP per capita and energy consumption per capita. Across the globe, the two metrics are positively correlated.

# Exploratory Data Analysis Part 2 - A Deep Dive on Nations Who Received Financial Aid for Clean Energy in the Years 2000-2019

## Calculating the total aid for clean energy by year, in dollars

```python
df_developing = df3[df3["Aid_for_Clean_Energy($)"].notnull()]
df_aid = df_developing[["Year","Aid_for_Clean_Energy($)"]].groupby("Year")
df_total_aid = df_aid.sum()

df_total_aid
```

## Pulling the averages by year of various clean energy metrics among nations who received funding:

```python
df4 = df_developing[["Year",
                     "Access_to_Electricity(%_Pop)",
                     "Renewable_Energy_Generating_Capacity/Capita",
                     "Percent_Clean_Energy",
                     "Energy_Consumption/Capita(kWh)",
                     "GDP_Per_Capita"]].groupby("Year")
df_developing_avgs = df4.mean()
df_developing_avgs
```

## Joining the two tables into one

```python
df_developing_by_yr = df_total_aid.merge(df_developing_avgs, left_index = True, right_index = True)
df_developing_by_yr.head()
```

## Visualizing metrics over time among countries who received funding

```python
fig, ax = plt.subplots()
ax.plot(df_developing_by_yr["Aid_for_Clean_Energy($)"])
ax.set_ylabel("Financial Aid for Clean Energy ($Billions)")
ax.legend(["Financial Aid"],loc="upper center")
ax.set_xlabel("Year")
ax.ticklabel_format(axis="y",scilimits=(9,9))


ax2 = ax.twinx()
ax2.plot(df_developing_by_yr["Access_to_Electricity(%_Pop)"], color = "magenta")
ax2.set_ylabel("Percent Population with Access to Electricity")
ax2.legend(["Electricity Access"],loc=2)


plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018])
plt.title("Clean Energy Aid & Electricity Access in Countries that Received Aid")
plt.rcParams["figure.figsize"] = (7,5)


plt.show()
```

Comments: Access to electricity in nations receiving financial aid increased nearly 20%.

```python
fig, ax = plt.subplots()
ax.plot(df_developing_by_yr["Aid_for_Clean_Energy($)"])
ax.set_ylabel("Financial Aid for Clean Energy ($Billions)")
ax.legend(["Financial Aid"],loc="upper center")
ax.set_xlabel("Year")
ax.ticklabel_format(axis="y",scilimits=(9,9))


ax2 = ax.twinx()
ax2.plot(df_developing_by_yr["GDP_Per_Capita"], color = "red")
ax2.set_ylabel("GDP per Capita (Dollars)")
ax2.legend(["GDP per Capita"],loc=2)


plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018])
plt.title("Clean Energy Aid & GDP per Capita in Nations that Received Aid")


plt.show()
```

Comments: GDP per capita in these nations tripled. This makes sense since most countries that routinely receive dollar funding are developing nations experiencing high population growth.

```python
fig, ax = plt.subplots()
ax.plot(df_developing_by_yr["Aid_for_Clean_Energy($)"])
ax.set_ylabel("Financial Aid for Clean Energy ($Billions)")
ax.legend(["Financial Aid"],loc=1)
ax.set_xlabel("Year")
ax.ticklabel_format(axis="y",scilimits=(9,9))


ax2 = ax.twinx()
ax2.plot(df_developing_by_yr["Percent_Clean_Energy"], color = "orange")
ax2.set_ylabel("Percent Clean Energy")
ax2.legend(["Clean Energy Share"],loc=2)


plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018])
plt.title("Clean Energy Aid & Clean Energy Share in Nations that Received Aid")
plt.rcParams["figure.figsize"] = (7,5)


plt.show()
```

Comments: The plot above shows that as financial aid for clean energy to developing nations has increased over the last 20 years, the percentage of clean energy out of total energy consumed in these countries has fallen. This paints a slightly dismal picture but there are many reasons this could be the case. For example, the US could intentionally be giving money to countries with lower clean energy share now than in 2000. Let's test this hypothesis.

## Updating data to only include nations that received funding every single year 2000-2019

```python
#Option 1 is to make a loop:

consistent_funding_list = list()
country_count = dict()

for country in df_developing["Country"]:
    country_count[country] = country_count.get(country, 0) + 1
    if country_count.get(country) == 20:
        consistent_funding_list.append(country)
    else:
        continue

print(consistent_funding_list)
```

```python
#Option 2 is to use the nifty value counts function:

df_developing["Country"].value_counts()
```

### There are many countries that appear less than 20 times. We will drop these countries and graph the data again.

```python
df_developing_adj = df_developing[df_developing["Country"].isin(consistent_funding_list)]
df_developing_adj.head()
```

```python
df_aid_adj = df_developing_adj[["Year","Aid_for_Clean_Energy($)"]].groupby("Year")
df_total_aid_adj = df_aid_adj.sum()

df_total_aid_adj.head()
```

```python
df4_adj = df_developing_adj[["Year",
                             "Access_to_Electricity(%_Pop)",
                             "Renewable_Energy_Generating_Capacity/Capita",
                             "Percent_Clean_Energy",
                             "Energy_Consumption/Capita(kWh)",
                             "GDP_Per_Capita"]].groupby("Year")
df_developing_avgs_adj = df4_adj.mean()
df_developing_avgs_adj.head()
```

```python
df_developing_by_yr_adj = df_total_aid_adj.merge(df_developing_avgs_adj, 
                                                 left_index = True, right_index = True)
df_developing_by_yr_adj.head()
```

## Updating the Clean Energy Funding vs Clean Energy Share visualization to only include countries who received funding every year 2000-2019

```python
fig, ax = plt.subplots()
ax.plot(df_developing_by_yr_adj["Aid_for_Clean_Energy($)"])
ax.set_ylabel("Financial Aid for Clean Energy ($Billions)")
ax.legend(["Financial Aid"],loc=1)
ax.set_xlabel("Year")
ax.ticklabel_format(axis="y",scilimits=(9,9))


ax2 = ax.twinx()
ax2.plot(df_developing_by_yr_adj["Percent_Clean_Energy"], color = "orange")
ax2.set_ylabel("Percent Clean Energy")
ax2.legend(["Clean Energy Share"],loc=2)


plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018])
plt.title("Clean Energy Aid & Clean Energy Share in Nations Receiving Aid")


plt.show()
```

Comments: Well, the downward trend in clean energy share is still apparent even in countries who have received funding for 20 years straight. The downward slope is even steeper and more consistent now.

## Could renewable energy generating capacity somehow be falling in countries who received financial aid for clean energy? 

```python
fig, ax = plt.subplots()
ax.plot(df_developing_by_yr_adj["Aid_for_Clean_Energy($)"])
ax.set_ylabel("Financial Aid for Clean Energy ($Billions)")
ax.legend(["Financial Aid"],loc="lower right")
ax.set_xlabel("Year")
ax.ticklabel_format(axis="y",scilimits=(9,9))


ax2 = ax.twinx()
ax2.plot(df_developing_by_yr_adj["Renewable_Energy_Generating_Capacity/Capita"], color = "lime")
ax2.set_ylabel("Renewable Energy Generating Capacity per Capita")
ax2.legend(["Renewable Energy Capacity"],loc=2)


plt.xticks([2000,2002,2004,2006,2008,2010,2012,2014,2016,2018])
plt.title("Clean Energy Aid & Clean Energy Capacity in Nations Receiving Aid")


plt.show()
```

Comments: Even though clean energy share has fallen in nations who continuously recevied funding since 2000, the renewable energy generating capacity per capita has exponentially increased, alongside the financial aid. This is a sign that financial aid money is indeed being used to increase renewable energy capacity, meaning the decrease in clean energy share is not a result of a decline in clean energy production.

The decline in clean energy share is more likely caused by unclean energy growing at a faster rate than clean energy, possibly to quickly support population growth in these developing nations.

# Conclusions

There is a correlation between the productivity/wealth of a country and their aversion to clean energy. With correlation we cannot infer causation, so more research will need to be conducted to determine if wealth leads countries to prefer unclean energy, or if clean energy stalls production and wealth generation. Another possibility is that certain countries may have become extremely wealthy by discovering large deposits of unclean energy (oil) within their borders. Oil and coal can be discovered whereas sunlight, wind, and nuclear energy need to be harnessed or built. The process of generating clean energy may simply be more difficult than the process of generating unclean energy, leading countries to relunctantly opt for unclean sources to support population growth and wealth creation.

Over the last two decades, clean energy share has fallen in countries who consistently received funding for clean energy production. However, financial aid for clean energy does not seem to be in vain and likely produces positive effects given the sharp increase in renewable energy generating capacity in these developing nations. It is likely that the fall in clean energy share comes from unclean energy growing at a faster rate than clean energy. Once again, in a rapidly growing country, it's easier to immediately use oil and coal than to harness the power of sunlight, wind, and nuclear energy. Over time, clean energy may catch up.
